```python
import math
S=1

def newtonExpandion(k,l,S):
    return (((-1)**(k-l)) * (math.comb(k,l)) * (math.sqrt(1-(l/(2*S)))) )/math.factorial(k)

forK1=[]

for S in range(1,100):
    print('S=',S)
    for k in range(int(2*S)+1):
        if k==1:
            print('    k=',k)
        
        expansion=0
        for l in range(k+1):
            expansion+=newtonExpandion(k,l,S)
        if k==1:
            forK1.append(expansion)
        print('          ',expansion)

```

```python
plt.plot(forK1)
plt.grid()
plt.show()
```
